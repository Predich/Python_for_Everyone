```python
filename  ="marathon_speed.csv"
my_file = open (filename,'r')

for each_line in my_file.readlines()[1:]:
    x = each_line.split(",")
    if float(x[1])>120:
        print(x)
my_file.close
```

```python
Opentowrite=open("marathon_speed_120.txt","w")
data="marathon_speed.csv"

# data=r"D:\python101\openfile\marathon_speed.csv"

opentoread=open(data,'r')
for each_line in opentoread.readlines()[1:]:
    x=each_line.split(",")
    if float(x[1])>120:
        Opentowrite.write(each_line)
        print(each_line)
data.close()
```
