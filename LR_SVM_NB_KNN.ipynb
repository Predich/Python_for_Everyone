## Example 1 : Diabetes(โรคเบาหวาน)

### Import Library

```python
import numpy as np

import pandas as pd

from sklearn.linear_model import LogisticRegression ## import Logistic Regression from Scikit-learn
```

### Read Data diabetes

```python
df_diabetes = pd.read_csv('../data set/diabetes.csv')
```

### Clear Data

```python
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']

df_diabetes.columns = col_names
```

```python
df_diabetes[['glucose','bp','skin','insulin','bmi']] = df_diabetes[['glucose','bp','skin','insulin','bmi']].replace(0,np.NaN)
```

```python
df_diabetes['glucose'].fillna(df_diabetes['glucose'].mean(), inplace = True)
df_diabetes['bp'].fillna(df_diabetes['bp'].mean(), inplace = True)
df_diabetes['skin'].fillna(df_diabetes['skin'].mean(), inplace = True)
df_diabetes['insulin'].fillna(df_diabetes['insulin'].mean(), inplace = True)
df_diabetes['bmi'].fillna(df_diabetes['bmi'].mean(), inplace = True)
```

```python
df_diabetes.isnull().sum()
```

### Feature Selection

```python
feature_cols = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp','pedigree']

x = df_diabetes[feature_cols] # Features

y = df_diabetes.label # Target variable
```

### Splitting Data

```python
from sklearn.model_selection import train_test_split # Import train_test_split function
```

```python
# Split dataset into training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

### Building Logistic Regression

```python
from sklearn.linear_model import LogisticRegression ## import Logistic Regression from Scikit-learn
```

```python
# instantiate the model (using the default parameters)
logreg = LogisticRegression(solver='lbfgs')  

## solver = Algorithm to use in the optimization problem

## lbfgs = Limited-memory Broyden–Fletcher–Goldfarb–Shanno algorithm
```

```python
# fit the model with data
logreg.fit(x_train,y_train)

#
y_pred=logreg.predict(x_test)
```

### Evaluating Model

```python
from sklearn import metrics

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
print(metrics.classification_report(y_test, y_pred))
```

# Support Vector Machine(SVM)

https://www.datacamp.com/community/tutorials/svm-classification-scikit-learn-python



```python
feature_cols = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp','pedigree']

x = df_diabetes[feature_cols] # Features

y = df_diabetes.label # Target variable

# Split dataset into training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

### Building Support Vector Machine(SVM)

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
svm = svm.SVC(kernel='linear') # linear , poly, rbf, sigmoid
```

```python
#Train the model using the training sets
svm.fit(x_train, y_train)

#Predict the response for test dataset
y_pred = svm.predict(x_test)
```

### Evaluating Model

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
print(metrics.classification_report(y_test, y_pred))
```

## Other Kernel

C = เป็นตัวกำหนดขนาดของ margin ของเส้น hyperplane

gamma = คือส่วนกลับของอัตราการลดค่าตามระยะห่างจากใจกลาง



```python
from sklearn import svm

#Create a svm Classifier

svm_2 = svm.SVC(kernel='poly', C = 1, gamma = 1 ) # linear , poly, rbf, sigmoid

#Train the model using the training sets
svm_2.fit(x_train, y_train)

#Predict the response for test dataset
y_2_pred = svm_2.predict(x_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_2_pred))
```

## Example 2 : IRIS 

## Import Library

```python
import numpy as np

import pandas as pd

from sklearn.ensemble import RandomForestClassifier ## import Random Forest from Scikit-learn

from sklearn.linear_model import LogisticRegression ## import Logistic Regression from Scikit-learn

from sklearn.naive_bayes import GaussianNB ## import Naive Bayes from Scikit-learn

from sklearn.neighbors import KNeighborsClassifier ## import K-Nearest Neighbor from Scikit-learn

from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation

from sklearn.model_selection import train_test_split # Import train_test_split function

import matplotlib.pyplot as plt #Import matplotlib for plot graph

import seaborn as sns #Import seaborn for plot graph
```

## Read Data

sepal length = ความยาวของกลีบเลี้ยง

sepal width = ความกว้างของกลีบเลี้ยง

petal length = ความยาวของกลีบดอกไม้

petal width = ความกว้างของกลีบดอกไม้

species = สายพันธุ์

```python
df_iris = pd.read_csv('../data set/iris.csv')
```

```python
col_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

df_iris.columns = col_names
```

```python
df_iris.isnull().sum()
```

## Correlation

```python
sns.set_context("notebook", font_scale=1.0, rc={"lines.linewidth": 2.5})

plt.figure(figsize=(10,5))

# create a mask so we only see the correlation values once
mask = np.zeros_like(corr)

mask[np.triu_indices_from(mask, 1)] = True

a = sns.heatmap(corr,mask=mask, annot=True, fmt='.2f')

rotx = a.set_xticklabels(a.get_xticklabels(), rotation=45)

roty = a.set_yticklabels(a.get_yticklabels(), rotation=30)
```

```python
# Boxplot

sns.boxplot(data = df_iris)
```

## Split Data

```python
feature_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

x = df_iris[feature_cols] # Features

y = df_iris.species # Target variable

# Split dataset into training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

## Build Model

```python
random_forest = RandomForestClassifier(criterion="entropy",n_estimators = 500)

random_forest.fit(x_train,y_train)

random_forest_pred = random_forest.predict(x_test)
```

```python
log_reg = LogisticRegression(solver='lbfgs')

log_reg.fit(x_train,y_train)

log_reg_pred = log_reg.predict(x_test)
```

```python
from sklearn import svm ## import Support Vector Machine from Scikit-learn

svm = svm.SVC(kernel='linear') # linear , poly, rbf, sigmoid

svm.fit(x_train,y_train)

svm_pred = svm.predict(x_test)
```

```python
nb = GaussianNB()

nb.fit(x_train,y_train)

nb_pred = nb.predict(x_test)
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)

#Train the model using the training sets
knn.fit(x_train, y_train)

#Predict the response for test dataset
knn_pred = knn.predict(x_test)
```

```python
print("Random Forest Accuracy : %.2f %%" %(metrics.accuracy_score(y_test, random_forest_pred)))

print("Logistic Regression Accuracy :  %.2f %%" %(metrics.accuracy_score(y_test, log_reg_pred)))

print("Support Vector Machine Accuracy :  %.2f %%" %(metrics.accuracy_score(y_test, svm_pred)))

print("Naive Bayes Accuracy :  %.2f %%" %(metrics.accuracy_score(y_test, nb_pred)))

print("K-nearest Neighbor :  %.2f %%" %(metrics.accuracy_score(y_test, knn_pred)))
```
